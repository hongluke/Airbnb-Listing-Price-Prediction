```python
#used sns.displot(ab.column) to see the distribution of data of numeric data

#use 1.0 (median) to replace the 200 missing values in bathrooms because right skewed distribution
ab["bathrooms"].fillna(ab["bathrooms"].median(skipna=True), inplace=True)   

#***use _____ to replace the 15827 missing values in first_review 
ab["first_review"].fillna(0)  

#use t (mode) to replace the 188 missing values in host_has_profile_pic because all of the rows are t (most frequent)
ab["host_has_profile_pic"].fillna(ab['host_has_profile_pic'].value_counts().idxmax(), inplace=True)

#use t (mode) to replace the 188 missing values in host_identity_verfied because occurs most frequently (about twice the amount of f)
ab["host_identity_verified"].fillna(ab['host_identity_verified'].value_counts().idxmax(), inplace=True)

#use 100% (mode) to replace the 18299 missing values in host_response_rate because occurs the most frequently (43252 times)
ab["host_response_rate"].fillna(ab['host_response_rate'].value_counts().idxmax(), inplace=True)

#***use _____ to replace the 188 missing values in host_since
ab["host_since"].fillna(0)  

#***use _____ to replace the 15827 missing values in last_review
ab["last_review"].fillna(0)  

#***use _____ to replace the 6872 missing values in neighbourhood 
ab["neighbourhood"].fillna(0)  

#use 100.0 (mode) to replace the 16722 missing values in review_scores_rating because most frequently appearing 
ab["review_scores_rating"].fillna(ab['review_scores_rating'].value_counts().idxmax(), inplace=True)

#***use _____ to replace the 966 missing values in zipcode 
ab["zipcode"].fillna(0)  

#use 1.0(median) to replace 91 missing values in bedrooms because right skewed distribution
ab["bedrooms"].fillna(ab["bedrooms"].median(skipna=True), inplace=True)  

#use 1.0(median) to replace 131 missing values in beds because right skewed distribution
ab["beds"].fillna(ab["beds"].median(skipna=True), inplace=True)  

ab.isnull().sum()  # check for missing value
```
